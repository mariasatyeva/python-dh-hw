### Задание 1
Дан набор слов "абракадабра фокус покус десятка за курс немедленно появись". Необходимо написать функцию, которая выделяет все символьные биграммы в этом наборе слов и возвращает их список. Символьные биграммы -- это пары стоящих рядом букв, например, "ра", "ку". (Но это не буквы разделенные пробелами или пара из буквы и пробела!) Выведите (применить функцию print) этот список. Создать словарь, где ключами будут полученные биграммы, а значениями — их количество в во введенном тексте. Проитерироваться по словарю, пока не встретится биграмма из двух одинаковых букв. На каждой итерации надо вывести (применить функцию print) рассматриваемую биграмму, и количество раз, которое она встретилось.

```python
def bigrams(text):
    textwords = text.split()
    letters = list(''.join(textwords))
    return [(letters[i], letters[i+1]) for i in range(0, len(letters)-1)]

bigram = bigrams("абракадабра фокус покус десятка за курс немедленно появись")
print (bigram) # вывести список

bdict = dict((i, bigram.count(i)) for i in bigram) # создание словаря

for i in bdict:    # итерация
    if str(list(i)[::-1]) != str(list(i)):  # увы, так как словарь - неупорядоченная система -
        print("Биграммы с разными буквами: " + str(i) + ". Количество раз в тексте: " + str(bigram.count(i)))
    else:   # я не придумала как остановить цикл на биграмме из одинаковых букв!
        print("Биграмма с одинаковыми буквами: " + str(i) + ". Количество раз в тексте: " + str(bigram.count(i)))
        # я нашла в интернете OrderDict, но скачивать его наверно было бы читерством :)
```

### Задание 2
Даны два слова, ваша программа должна проверять является ли второе слово обратной записью первого слова.

```python
def words(word1, word2):
    list1 = list(word1)[::-1]
    list2 = list(word2)
    if str(list2) == str(list1):
        return "YES"
    else:
        return "NO"
    
print(words("шрам", "марш"))
print(words("слабый", "сильный"))
```
